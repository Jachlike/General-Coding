```python
# A function to retrieve total and percentage of missing data in a dataset
def cal_missing_val(data):
    total = data.isnull().sum().sort_values(ascending=False)
    percent = (data.isnull().sum() / data.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat(
        [total, percent * 100], axis=1, keys=["Total", "Percent(%)"]
    )
    return missing_data

# Convert empty spaces to null values
gamedf.replace(r"^\s*$", np.nan, regex=True, inplace=True)
cal_missing_val(gamedf)
```

## The missing variables are considered 'object' data types and contain '?' values

### CHANGE THE '?' VALUES TO NULL NULL VALUES AND CHANGE THE DATA TYPE PF THE MISSING VARAIBLES

```python
##1 
invalid_dtypecol = ['Age', 'TotalHours', 'HoursPerWeek']

##2 convert '?' to null values
gamedf[invalid_dtypecol] = gamedf[invalid_dtypecol].apply(lambda x: x.replace("?", np.nan))

##3
for col in invalid_dtypecol:
    gamedf[col] = pd.to_numeric(gamedf[col])
```
